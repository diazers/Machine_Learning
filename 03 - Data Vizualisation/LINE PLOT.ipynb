```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100) # membuat array uhtuk data kita
y = np.sqrt(x)
```

## Inisialisasi Figure dan Axes

figure adalah bidang untuk kita menaruh diagram kita  
axes adalah sumbu pada figure kita

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6)) # mengatur ukuran figure kita
ax = plt.axes()
ax.plot(x,y)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# plot dua grafik dalam satu figure

fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1)
ax.plot(x, y2)
```

```python
fig = plt.figure(figsize=(6,6)) # cara cepat ploting mengggunakan plt
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r')  # mengatur warna garis
plt.plot(x, y2, 'b')
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r--') # mengatur pola garis
plt.plot(x, y2, 'b-.')
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'ro') # mnegatur pola garis
plt.plot(x, y2, 'bv')
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r-', linewidth=5) # mengatur tebal garis
plt.plot(x, y2, 'bv', markersize=3) # mengatur tebal marker
plt.ylim(-2, 2)
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r-', linewidth=5)
plt.plot(x, y2, 'bv', markersize=3)
plt.ylim(-2, 2) # mengatur skala sumbu x
plt.xlim(0, 20) # mengatur skala sumbu y
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r-', linewidth=5)
plt.plot(x, y2, 'bv', markersize=3)
plt.ylim(-2, 2)
plt.xlim(0, 10)
plt.axis('equal') # membuat sumbu x dan y menjadi skala dengan interval yang sama

# versi shortcut menggunakan plt
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5)
ax.plot(x, y2, 'bv', markersize=3)
ax.set_ylim(-2, 2)
ax.set_xlim(0, 10)
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5)
ax.plot(x, y2, 'bv', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2)) # dibuat lebih ringkas 
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r-', linewidth=5)
plt.plot(x, y2, 'bv', markersize=3)
plt.ylim(-2, 2)
plt.xlim(0, 10)
plt.axis('equal')
plt.title("sin dan cos")
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5)
ax.plot(x, y2, 'bv', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2))
ax.set_title("sin dan cos")
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5)
ax.plot(x, y2, 'bv', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos") # lebih ringkas
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r-', linewidth=5)
plt.plot(x, y2, 'bv', markersize=3)
plt.ylim(-2, 2)
plt.xlim(0, 10)
plt.axis('equal')
plt.title("sin dan cos")
plt.xlabel('x') # menambah label pada sumbu x
plt.ylabel('y') # menambah label pada sumbu y
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5)
ax.plot(x, y2, 'bv', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5)
ax.plot(x, y2, 'bv', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2))
ax.set_title("sin dan cos", fontsize=20) # model penulisan parameter terpisah bisa lebih fleksibel menentulkan styling
ax.set_xlabel('x' , fontsize=10)
ax.set_ylabel('y' , fontsize=10)
ax.axis('equal'); # tanda ';' digunakan agar tidak mengeluarkan hasil angka di bagan awal atau atas dari plot
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5, label='sin')
ax.plot(x, y2, 'bv', markersize=3, label='cos')
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axis('equal')
ax.legend() # menambahkan legend pada plot
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(x, y1, 'r-', linewidth=5, label='sin')
ax.plot(x, y2, 'bv', markersize=3, label='cos')
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axis('equal')
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1) # styling legend
```
