# SUBPLOT

memplot banyak grafik dalam figure yang sama

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numbers = np.random.randn(10000)
```

```python
df = pd.read_csv('data/circle_data.csv', sep=';') # delimiter atau separator terganmtung data raw
df.head()

x = df.x
y = df.y
rad = df.radius
```

```python
plt.subplot(231) # 2 baris 3 kolom urutan ke 1
plt.hist(numbers, bins=80, density=1);

plt.subplot(232) # 2 baris 3 kolom urutan ke 2
plt.hist(numbers, bins=80, density=1);

plt.subplot(235) # 2 baris 3 kolom urutan ke 5
plt.scatter(x, y, c='r', s=10) # merubah warna
plt.axis("equal");

plt.subplot(236) # 2 baris 3 kolom urutan ke 6
plt.hist(numbers, bins=80, density=1);
```

```python
plt.figure(figsize=(15 , 10))

plt.subplot(2,3,1) # 2 baris 3 kolom urutan ke 1
plt.hist(numbers, bins=80, density=1);

plt.subplot(2,3,2) # 2 baris 3 kolom urutan ke 2
plt.hist(numbers, bins=80, density=1);

plt.subplot(2,3,5) # 2 baris 3 kolom urutan ke 5
plt.scatter(x, y, c='r', s=10) # merubah warna
plt.axis("equal");

plt.subplot(2,3,6) # 2 baris 3 kolom urutan ke 6
plt.hist(numbers, bins=80, density=1);
```

cara yang berbeda

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))
ax[0,0].hist(numbers, bins=80, density=1);
ax[0,2].hist(numbers, bins=80, density=1);
ax[1,1].scatter(x, y, c='r', s=10) # merubah warna
ax[1,1].axis("equal");
ax[1,2].hist(numbers, bins=80, density=1);
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10), sharex=True) # skala sumbu x hanya menggunakan satu skala dan berlaku ke semua plot
ax[0,0].hist(numbers, bins=80, density=1);
ax[0,2].hist(numbers, bins=80, density=1);
ax[1,1].scatter(x, y, c='r', s=10) # merubah warna
ax[1,1].axis("equal");
ax[1,2].hist(numbers, bins=80, density=1);
```
