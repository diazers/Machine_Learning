```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8))  # membuat bidang 3D
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

x, y = np.meshgrid(x,y)
z = np.sqrt(x**2 + y**2 + 10)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False, cmap='winter')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, antialiased=False, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10, azim=45)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(x, y, z, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10, azim=45)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z, 50, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10, azim=45)
```
