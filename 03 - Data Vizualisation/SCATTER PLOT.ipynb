# Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/circle_data.csv', sep=';') # delimiter atau separator terganmtung data raw
df.head()
```

```python
x = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.axis("equal");
```

```python
plt.scatter(x, y, c='r') # merubah warna
plt.axis("equal");
```

```python
plt.scatter(x, y, c='r', s=10) # merubah size
plt.axis("equal");
```

```python
plt.scatter(x, y, c='r', s=rad*50) # merubah size berdasarkan radius, makin ke pusat makih kecil
plt.axis("equal");
```

```python
plt.scatter(x, y, c=rad, s=rad*50, cmap='seismic')
plt.axis("equal");
```
