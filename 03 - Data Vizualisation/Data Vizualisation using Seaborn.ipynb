```python
df = pd.read_csv("data/cereal.csv", index_col="name")
df.head()
```

## Misiing value Vizualisation

```python
plt.figure(figsize=(6, 6))
sns.heatmap(df.isna(), cmap="Blues", cbar=False)

# memperlihatkan data yang hiang
```

```python
plt.figure(figsize=(6, 6))
sns.heatmap(~df.isna(), cmap="Blues", cbar=False)

# memperlihatkan data yang tidak hil|ang
```

## Correlation Heatmap

```python
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), cmap="bwr", vmin=-1, vmax=1, annot=True)
```

## PairPlot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'])

# pisahkan berdasrkan lokasi rak dan melihat korelasi dengan kalori lemak dan rating
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'], kind="reg")

# pisahkan berdasrkan lokasi rak dan melihat korelasi dengan kalori lemak dan rating
# juga menambahkan garis regresi
```

```python
sns.jointplot('potass', 'fiber', data=df, kind='reg')

# menggabungkan dua kolom ke dalam scaterplot dan histogram
```
