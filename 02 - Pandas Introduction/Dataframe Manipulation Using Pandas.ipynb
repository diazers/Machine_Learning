# Dataframe Manipulation Using Pandas
---

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/cereal_imputed.csv', index_col='name')
df.head()
```

## Filtering

conditional formating sama seperti banyak bahasa pemrograman menggunakan operasi aritmatika dan perbandingan (+, -, /, x) (<, >, ==, |, ~) etc

```python
df[df.rating > 70]
# menampilkan cereal dengan rating diatas 70
```

```python
df[~(df.shelf == 1)]
# menampilkan shelf  bernilai selain 1 karena ada tanda negasi yang berarti selain
```

```python
# menampilkan beberapa kondisi filtering sekaligus menggunakan (&)
# & akan menampilkan hasil yang memenuhi dua kondisi
df[(df.rating > 50) & (df.fat == 0)]
```

```python
# menampilkan beberapa kondisi filtering sekaligus menggunakan (|) artinya atau
# & akan menampilkan hasil yang memenuhi salah satu kondisi atau dua kondisi
df[(df.rating > 70) | (df.vitamins == "FDA_100")]
```

```python
df[df.mfr.isin(['K', 'N']) & (df.rating > 70)]

# isin digunakan untuk menampilkan salah satu atau beberapa kategori dari suatu kolom
```

```python
# membuat result agar tidak mempengaruhi dataframe raw / asli bisa menggunakan .copy()

df_my_preferences = df[(df.rating > 60) & (df.mfr.isin(['K', 'N']))].copy()
df_my_preferences
```

## Sorting

```python
# sorting / mengurutkan berdasarkan kolom tertentu

df_my_preferences.sort_values('rating')

# akan mengurutkan dari kecil ke besar (Ascending)
```

```python
df_my_preferences.sort_values('rating', ascending=False)

# akan mengurutkan dari besar ke kecil (Descending)
```

```python
df_my_preferences.sort_values('rating', ascending=False, inplace=True)
df_my_preferences
# gunakan parameter inplace=True untuk membuat perubahan ke data asli
```

## Grouping and Aggregate

```python
df.groupby("shelf").mean()

# digroup berdasarkan kategori yang ada di kolom shelf dan kolom lain diaggregate berdasarkan mean atau di rata-rata
# kolom yang lain yang tidak berbentuk numerik tidak bisa diagregate dan tidak akan ditampilkan
```

```python
# groupby lebih dari satu kolom

df.groupby(['mfr', 'shelf']).mean()

# akan digrouping berdasarkan kategori yang ada di kolom mfr dan shelf
```

```python
df.groupby(['mfr', 'shelf']).count()

# groupby lebih dari satu kolom dan agregate berdasarkan hitung
```

```python
df.groupby(['mfr', 'shelf']).size()

# melihat jumlah secara keseluruhan
```

```python
# pivot table

df.pivot_table(index='mfr', columns='vitamins', values='rating', aggfunc='mean')
```

## Map, Apply, dan Transform

### MAP

```python
df['shelf_name'] = df.shelf.map({1:"bawah" , 2:"tengah" , 3:"atas"})
df.head()

# membuat kolom baru dan memasukan nilai baru dan dipetakan berdasarkan koom tertentu 
```

```python
df.shelf = df.shelf.map({1:"bawah" , 2:"tengah" , 3:"atas"})
df.head()

# membuat kolom baru dan menimpa kolom lama dengan nilai baru dan dipetakan berdasarkan koom tertentu 
```

### Apply

```python
def convert_g_to_mg(x):
    return 1000*x

def convert_to_lower(x):
    return x.lower()
```

```python
df['fat_mg'] = df.fat.apply(convert_g_to_mg)
df.head()
```

```python
df['mfr_lower'] = df.mfr.apply(convert_to_lower)
df.head()
```

```python
df['protein_mg'] = df.protein.apply(lambda x: x*1000)
df.head()

# menggunakan fungsi sekali pakai (disposable) menggunakan lambda function
```

### Transform

```python
df['scaled_rating'] = df.rating.transform(lambda x: (x-x.min())/(x.max()-x.min()))
df.head()

# scaling minmax cara manual
```

```python
df.drop(columns=['shelf_name', 'fat_mg', 'mfr_lower', 'protein_mg', 'scaled_rating'], inplace=True)
df.head()
```

## ENCODING DAN BINNING

konsep encoding
<img src="picture/konsep_encoding.png">

mengubah data kategorikal (tipe data object/string) pada suatu kolom menjadi numerikal (angka)

label encoding / ordinal encoding pada dasarnya sama seperti pada maping yang telah dilakukan sebelumnya, umumnya dilakukan apabila kolom ketegorikal yang berbentuk binary (dua pilihan). kita mengkodifikasi objek menjadi numerikal value

bila lebih dari dua kategori umumnya menggunakan one hot encoding atau dummy variabel/ one hot encoding. akan dibagi menjadi kolom-kolom baru berdasarkan kategori


### ONE HOT ENCODING / DUMMY VARIABLE

```python
pd.get_dummies(df, columns=['vitamins'])

# kategori pada kolom vitamins akan dibelah dqan dibuat menjadi kolom baru terpisah
```

### Binning

mengubah data numerik menjadi kategori, misal data numerik akan dikelompokan berdasarkan interval tertentu dan menjadi beberapa grup

contoh :
- 0-5 = balita
- 5-10 = anak-anak
- 10-18 = remaja

```python
df.sodium.hist(bins=10)
```

```python
df['sodium_bin'] = pd.cut(df.sodium, bins=3, labels=['low', 'medium', 'high'])
df.head()

# membagi kolom sodium menjadi 3 kategori low, medium, dan high
```

```python
df['sodium_bin'] = pd.cut(df.sodium, bins=[0, 100, 200, df.sodium.max()], labels=['0-100', '100-200', '>200'])
df.head()

# membagi berdasarkan angka interval yang kita tentukan
# 0-100
# 100-200
# 200-max
```

```python
df.rating.hist(bins=50)
```

```python
df['Grade'] = pd.cut(df.rating, bins=[0, 40, 70, df.rating.max()], labels=['C', 'B', 'A'])
df.head()
```
