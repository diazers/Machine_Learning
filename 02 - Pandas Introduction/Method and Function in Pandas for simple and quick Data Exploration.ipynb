### mengetahui data hilang (Missing Value)/Nan

```python
df.isna().sum()
# memperlihatkan data hilang untuk tiap kolom 
```

### Drop missing value

```python
df.dropna()

#menghapus baris data yang hilang, terlihat jumlah baris data berkurang
```

### Impute Missing Value

```python
df.fillna(df.mean())

# manampilkan data yang hilang menjadi rata-ratanya (mean)
```

```python
df.fillna(df.mean(), inplace=True)

# inplace=True akan membuat data yang hilang pada data asli akan diisi oleh rata-rata
```

### menyimpan dataframe menjadi CSV

```python
df.to_csv("data/cereal_imputed.csv", index_label='name')

# menyimpan datframe yang sudah dianalisis menjadi csv kembali
```
